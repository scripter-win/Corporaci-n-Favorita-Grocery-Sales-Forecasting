```python
import math
def deviation(data):
    mean = data.mean()
    var = 0.0
    for i in data:
        var += (mean-i)**2
    return math.sqrt(var/len(data))
    
    
print(len(df['brand_name']))
sets = set([i for i in df.brand_name])
print(len(sets))
print(set)
for i in sets:
    data = df[df.brand_name==i].price
    if(data.empty):
        continue
    print('item:',i,' mean: ',df[df.brand_name==i].price.mean(),' deviation: ',deviation(df[df.brand_name==i].price))
```
